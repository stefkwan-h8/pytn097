```python
# selain bisa import data, dan baca data per baris/kolom
# kita bisa ambil data dengan kondisi tertentu

df_gold.head()
```

```python
df_gold.shape
```

```python
# misalkan kita mau ambil data khusus yang volume diatas 100

# kondisi adalah data yang ada di kolom Volume, nilainya lebih dari 100
kondisi = df_gold["Volume"] > 100
# ini menjadi kumpulan boolean

# kumpulan boolean kondisi itu kita gunakan sebagai index, sebagai filter
df_gold[kondisi]

# teknik ini juga disebut sebagai "query"
```

```python
# ditulis langsung juga bisa

df_gold[df_gold["Volume"] > 100]
```

```python
# kita juga bisa mengambil data dengan lebih dari 1 kondisi
# misalkan kita mau data harga stok emas di tahun 2000 dan di bulan 3 saja

# untuk memudahkan pembuatan kondisi, kita tambahkan 2 kolom baru: Tahun dan Bulan
df_gold["Date"]
```

```python
df_gold["Date"].dt.year # kita masukin ke kolom baru dengan judul "Tahun"
```

```python
df_gold["Date"].dt.month
```

```python
# nambah kolom baru, kasih judul, kasih datanya
df_gold["Tahun"] = df_gold["Date"].dt.year
df_gold["Bulan"] = df_gold["Date"].dt.month

df_gold.head()
```

```python
# kita buat 2 kondisi: Tahun 2000, Bulan 3

kondisi1 = df_gold["Tahun"] == 2000
kondisi2 = df_gold["Bulan"] == 3

# dua kondisi, kita mau dua2nya harus True, kita pakai "&" and

kondisi = kondisi1 & kondisi2

df_gold[kondisi]
```

```python
# langsung juga boleh
df_gold[(df_gold['Tahun'] == 2000) & (df_gold["Bulan"] == 3)]
```

kalau kondisi salah satu saja yang benar juga OK, kita bisa pakai or `|`

```python
# misalkan khusus yang Volume 100 atau 1000
df_gold[(df_gold['Volume'] == 100) | (df_gold["Volume"] == 1000)]
```

# Grouping & Aggregation

seringkali kita perlu grup dan agregat data sekaligus. Jarang kita perlu agregat saja.

contohnya, kita bisa cari rata2 harga stok emas dari data kita diatas. (agregat)

tapi seringali kita perlu cari rata2 harga stok emas per tahun. (group + agregat)

dengan grouping dan aggregation yang kita pakai bareng, kita bisa melihat trend, contohnya trend harga stok emas dari tahun ke tahun

```python
# lihat rata2 harga stok emas dari kolom Close untuk per tahun

df_gold.head()
```

```python
# agregasi, kolom Close, cari mean
df_gold["Close"].mean()
```

```python
# grouping per tahun, kolom Tahun
df_gold.groupby(["Tahun"])["Close"].mean()
```

```python
# visualisasi
df_gold_mean_per_tahun = df_gold.groupby(["Tahun"])["Close"].mean()

df_gold_mean_per_tahun.plot()
```

```python
df_gold_mean_per_tahun.plot(kind="bar")
```

# handling missing values
1. drop semua baris data yang ada missing value
2. drop kolom yang punya banyak missing value
3. isi missing value dengan suatu value tertentu, untuk contoh kali ini, yang tidak ada notes, kita tulis saja "no notes"

```python
# drop semua baris data yang ada missing value

df_nba_1 = df_nba.dropna()
df_nba_1.info()
```

```python
# drop kolom data yang ada missing value
df_nba_2 = df_nba.dropna(axis=1)
df_nba_2.info()
```

```python
# isi missing value
# biasanya isi per kolom
df_nba_3 = df_nba.copy()

df_nba_3["notes"] = df_nba_3["notes"].fillna("no notes")
df_nba_3.info()
```

```python
df_nba_3.head(10)
```

# invalid value
ada data yang tidak masuk akal di dataset kita.

case by case, misalkan untuk data game basket nasional Amerika NBA, aneh kalau ada game yang skor nya 0, atau skor dibawah 10.

bisa juga kita cari tren datanya terlebih dahulu pakai describe, atau teknik agregasi lainnya, sebelum lanjut cek invalid values

bisa juga tipe data tercampur, seperti tulisan "forfeit" di kolom pts yang harusnya isi angka semua.

```python
df_nba.describe()
```

```python
# di kasus ini, kita perhatikan keanehan pts dengan value 0
# kita gunakan teknik filter/ query, untuk cari data mana yang kolom pts punya value 0
kondisi = df_nba["pts"] == 0

df_nba[kondisi]
```

```python
kondisi = df_nba["pts"] == 2

df_nba[kondisi]
```

opsi handling invalid values:
- kita bisa hapus
- kita bisa edit jadi value yang normal
- kita bisa note secara terpisah dan dibiarkan

```python
# hapus
# hapus 1 baris menggunakan index
df_nba_4 = df_nba.drop(index=[26684])

kondisi = df_nba_4["pts"] == 0
df_nba_4[kondisi]
```

kita bisa kasih note seperti ini, kalau ada 1 game yang forfeit yang pts nya 0 dan 2, tapi kita biarkan di dalam dataset.

```python
# edit jadi value normal
df_nba_5 = df_nba.copy()

df_nba_5.loc[26684, "pts"] = 100

df_nba_5.describe()
```
